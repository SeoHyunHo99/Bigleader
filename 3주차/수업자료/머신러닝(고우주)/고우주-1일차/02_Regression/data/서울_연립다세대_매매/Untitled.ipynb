```python
import os
import pandas as pd 
path = 'C:/Users/shh28/BigLeader/3주차/수업자료/머신러닝(고우주)/고우주-1일차/02_Regression/data/서울_연립다세대_매매'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')]

file_list_py
```
