```python
#Step 5. 자동으로 검색어 입력 후 조회하기
element = driver.find_element_by_id("query")
driver.find_element_by_id("query").click( )
element.send_keys(query_txt)
element.send_keys("\n")
time.sleep(2)

#Step 6.학위 논문 선택하기
driver.find_element_by_link_text('학위논문').click()
time.sleep(2)
```

```python
#Step 7.Beautiful Soup 로 본문 내용만 추출하기
from bs4 import BeautifulSoup
html_1 = driver.page_source    # 
soup_1 = BeautifulSoup(html_1, 'html.parser')

content_1 = soup_1.find('div','srchResultListW').find_all('li')
for i in content_1 :
    print(i.get_text().replace("\n",""))
```

```python
#Step 8. 총 검색 건수를 보여주고 수집할 건수 입력받기
import math
total_cnt = soup_1.find('div','searchBox pd').find('span','num').get_text()
print('검색하신 키워드 %s (으)로 총 %s 건의 학위논문이 검색되었습니다' %(query_txt,total_cnt))
collect_cnt = int(input('이 중에서 몇 건을 수집하시겠습니까?: '))
collect_page_cnt = math.ceil(collect_cnt / 10)
print('%s 건의 데이터를 수집하기 위해 %s 페이지의 게시물을 조회합니다.' %(collect_cnt,collect_page_cnt))
```

```python
#Step 9. 각 항목별로 데이터를 추출하여 리스트에 저장하기
no2 = [ ]        #번호 저장
title2 = [ ]     #논문제목 저장
writer2 = [ ]    #논문저자 저장
org2 = [ ]       #소속기관 저장
no = 1

# 다음 페이지 번호 만들기
page_no=[ ]

for i in range(10,collect_cnt) :
    if i % 10 == 0 :
            page_no.append(i + 1)

for a in range(1, collect_page_cnt + 1) :
    
    html_2 = driver.page_source
    soup_2 = BeautifulSoup(html_2, 'html.parser')

    content_2 = soup_2.find('div','srchResultListW').find_all('li')
    
    for b in content_2 :    
        #1. 논문제목 있을 경우만
        try :
            title = b.find('div','cont').find('p','title').get_text()
        except :
            continue
        else :
            f = open(ft_name, 'a' , encoding="UTF-8")
            print('1.번호:',no)
            no2.append(no)
            f.write('\n'+'1.번호:' + str(no))

            print('2.논문제목:',title)
            title2.append(title)
            f.write('\n' + '2.논문제목:' + title)
            
            writer = b.find('span','writer').get_text()
            print('3.저자:',writer)
            writer2.append(writer)
            f.write('\n' + '3.저자:' + writer)

            org = b.find('span','assigned').get_text()
            print('4.소속기관:' , org)
            org2.append(org)
            f.write('\n' + '4.소속기관:' + org + '\n')
            
            f.close( )
            
            no += 1
            print("\n")
            
            if no > collect_cnt :
                break

            time.sleep(1)        # 페이지 변경 전 1초 대기 

        c = math.floor(a/10)-1 
        a += 1

#        if a == page_no[c]:
#            driver.find_element_by_link_text('다음 페이지로').click()
#        else :
#            driver.find_element_by_link_text('%s' %a).click() # 다음 페이지번호 클릭

        try :
            driver.find_element_by_link_text("%s" %a).click()
        except :
            driver.find_element_by_link_text("다음 페이지로").click()

print("요청하신 작업이 모두 완료되었습니다")
```

```python
# Step 10. 수집된 데이터를 xls와 csv 형태로 저장하기
import pandas as pd 

df = pd.DataFrame()
df['번호']=no2
df['제목']=pd.Series(title2)
df['저자']=pd.Series(writer2)
df['소속(발행)기관']=pd.Series(org2)

# xls 형태로 저장하기
df.to_excel(fx_name,index=False, encoding="utf-8")

# csv 형태로 저장하기
df.to_csv(fc_name,index=False, encoding="utf-8-sig")

print('요청하신 데이터 수집 작업이 정상적으로 완료되었습니다')
```

```python
# 네이버 블로그 정보 수집하여 저장하기

print("=" *80)
print(" 연습문제 : 저장할 내용을 목록으로 만들어서 xls , csv 형식으로 저장하기")
print("=" *80)

#Step 0. 필요한 모듈과 라이브러리를 로딩하고 검색어를 입력 받습니다

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys        # system 설정을 변경하기 위해 필요합니다
import math
import pandas  as pd    

query_txt = input('1.크롤링할 키워드는 무엇입니까?: ')
cnt=int(input('2.수집할 데이터는 몇 건입니까?: ') )
page_cnt = math.ceil(cnt / 30)

fc_name=input('3.csv 형태로 저장할 경로와 파일명을 확장자 포함해서 쓰세요(예:c:\\temp\\naver.csv): ')
fx_name=input('4.xls 형태로 저장할 경로와 파일명을 확장자 포함해서 쓰세요(예:c:\\temp\\naver.xls): ')

#Step 1. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.
path = "c:/temp/chromedriver_91/chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get('http://www.naver.com')
time.sleep(2)

#Step 2. 네이버 검색창에 입력 받은 검색어를 넣고 검색한 후 "블로그" 선택
element = driver.find_element_by_id("query")
element.send_keys(query_txt)
element.submit()

driver.find_element_by_link_text("VIEW").click()
time.sleep(2)

driver.find_element_by_link_text("블로그").click()

# Step 9. 저장 목록을 만든 후 목록에 있는 내용을 파일에 저장하기

no2 = [ ]           # 게시글 번호 컬럼
title2 = [ ]        # 게시물 제목 컬럼
contents2 = [ ]     # 게시글 내용 컬럼
bdate2 = [ ]        # 작성 일자 컬럼
nick2 = [ ]         # 블로그 닉네임

#Step 5. 현재 페이지의 내용을 저장 목록을 만든 후 목록에 있는 내용을 파일에 저장하기
no = 1

# 자동 스크롤다운 함수
def scroll_down(driver):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
  time.sleep(8)

i = 1
while (i <= page_cnt):
    scroll_down(driver) 
    i += 1
    print('%s 페이지 정보를 추출하고 있으니 잠시만 기다려 주세요~~^^' %i)
    

print("\n")

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

view_list = soup.find('ul','lst_total').find_all('li')

for i in view_list :
    
    no2.append(no)                            # 게시물 번호 리스트에 추가
    print('1.번호:',no)

    all_title = i.find_all('a')
    title = all_title[2].get_text( )          # 게시물 제목
    title2.append(title)                      # 게시물 제목 리스트에 추가
    print('2.제목:',title)

    try :
        contents = i.find('div' , 'api_txt_lines dsc_txt').get_text( )   # 게시물 내용
    except :
        contents = '본문 요약 내용이 없습니다'
        contents2.append(contents)                # 게시물 내용 리스트에 추가
        print('3.요약내용:',contents)
    else :
        contents2.append(contents)                # 게시물 내용 리스트에 추가
        print('3.요약내용:',contents)

    bdate = i.find('span','sub_time sub_txt').get_text( )  # 작성일자
    bdate2.append(bdate)                     # 작성일자 리스트에 추가
    print('4.작성일자:',bdate)

    print("\n")

    if no == cnt :
        break
        
    no += 1

# 출력 결과를 표(데이터 프레임) 형태로 만들기

naver_blog = pd.DataFrame()
naver_blog['번호'] = no2
naver_blog['제목'] = title2
naver_blog['요약내용'] = contents2
naver_blog['작성일자'] = bdate2

# csv 형태로 저장하기
naver_blog.to_csv(fc_name,encoding="utf-8-sig",index=False)
print(" csv 파일 저장 경로: %s" %fc_name) 

# 엑셀 형태로 저장하기
naver_blog.to_excel(fx_name , index=False)
print(" xls 파일 저장 경로: %s" %fx_name) 
```

```python
# 구 주소 블로그 : https://blog.naver.com/junhyuk_abba/221286810022
# 신 주소 블로그 : https://blog.naver.com/suheeryu/221314766979

print("=" *80)
print(" 연습문제: 블로그  크롤러 : 네이버 view -> 블로그 정보 수집하기")
print("=" *80)

#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas  as pd    
import math

#Step 2. 필요한 정보 입력받기
query_txt = input('1.정보를 수집할 키워드는 무엇입니까?: ')
start_date = input('2.조회를 시작할 날짜를 입력하세요(예:2017-01-01) :')
end_date = input('3.조회를 종료할 날짜를 입력하세요(예:2017-12-31): ')
cnt = int(input('4.몇 건의 정보를 수집할까요? :'))
page_cnt = math.ceil( cnt / 30 )
f_dir = input("5.파일을 저장할 폴더명만 쓰세요(기본값:c:\\temp\\):")
if f_dir == '' :
    f_dir="c:\\temp\\"

# Step 3. 결과를 저장할 파일명을 지정하기
import os

n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' %(n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)   

os.makedirs(f_dir+'NAVER_VIEW'+'-'+s+'-'+query_txt)
 
ft_name = f_dir+'NAVER_View'+'-'+s+'-'+query_txt+'\\'+'NAVER_View'+'-'+s+'-'+query_txt+'.txt'
fc_name = f_dir+'NAVER_View'+'-'+s+'-'+query_txt+'\\'+'NAVER_View'+'-'+s+'-'+query_txt+'.csv'
fx_name = f_dir+'NAVER_View'+'-'+s+'-'+query_txt+'\\'+'NAVER_View'+'-'+s+'-'+query_txt+'.xls'    
print("\n")  

s_time = time.time( )

#Step 4. 검색어 입력한 후 검색하여 View로 이동하기
import chromedriver_autoinstaller
chromedriver_autoinstaller.install()
driver= webdriver.Chrome()

url = 'https://www.naver.com'
driver.get(url)
time.sleep(2)

driver.find_element_by_id('query').click()
ele = driver.find_element_by_id('query')
ele.send_keys(query_txt)
ele.send_keys("\n")
time.sleep(3)

driver.find_element_by_link_text('VIEW').click()
driver.find_element_by_link_text('블로그').click()
time.sleep(1)

# 옵션 -> 날짜 입력하기
driver.find_element_by_link_text('옵션').click()
time.sleep(1)

# 옵션 - 날짜 - 직접입력 클릭
driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[1]/a[9]').click()
time.sleep(0.5)

# 조회 시작 날짜와 종료 날짜 선택
start_date2 = start_date.split('-')
start_year = start_date2[0]
start_mon = str(int(start_date2[1]))
start_day = str(int(start_date2[2]))

end_date2 = end_date.split('-')
end_year = end_date2[0]
end_mon = str(int(end_date2[1]))
end_day = str(int(end_date2[2]))

#시작 날짜 부분 클릭
driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[1]/span[1]/a').click()

#시작 년도 클릭
driver.find_element_by_link_text(start_year).click()
time.sleep(1)

#시작 월의 값은 xpath 로 선택합니다
if start_mon == '1' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[1]/a').click()
    time.sleep(1)
elif start_mon == '2' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[2]/a').click()
    time.sleep(1)
elif start_mon == '3' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[3]/a').click()
    time.sleep(1)
elif start_mon == '4' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[4]/a').click()
    time.sleep(1)
elif start_mon == '5' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[5]/a').click()
    time.sleep(1)
elif start_mon == '6' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[6]/a').click()
    time.sleep(1)
elif start_mon == '7' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[7]/a').click()
    time.sleep(1)
elif start_mon == '8' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[8]/a').click()
    time.sleep(1)
elif start_mon == '9' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[9]/a').click()
    time.sleep(1)
else :
    driver.find_element_by_link_text(start_mon).click()
    time.sleep(1)

#시작일의 값도 xpath 값을 사용합니다.
if start_day == '1' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[1]/a').click()
    time.sleep(1)                 
elif start_day == '2' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[2]/a').click()
    time.sleep(1)
elif start_day == '3' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[3]/a').click()
    time.sleep(1)
elif start_day == '4' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[4]/a').click()
    time.sleep(1)
elif start_day == '5' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[5]/a').click()
    time.sleep(1)
elif start_day == '6' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[6]/a').click()
    time.sleep(1)
elif start_day == '7' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[7]/a').click()
    time.sleep(1)
elif start_day == '8' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[8]/a').click()
    time.sleep(1)
elif start_day == '9' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[9]/a').click()
    time.sleep(1)
else :
    driver.find_element_by_link_text(start_day).click()
    time.sleep(1)

#종료 날짜 부분 클릭
driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[1]/span[3]/a').click()

#종료 년도 클릭
driver.find_element_by_link_text(end_year).click()
time.sleep(1)  

#종료 월의 값은 xpath 로 선택합니다
if end_mon == '1' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[1]/a').click()
    time.sleep(1)
elif end_mon == '2' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[2]/a').click()
    time.sleep(1)
elif end_mon == '3' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[3]/a').click()
    time.sleep(1)
elif end_mon == '4' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[4]/a').click()
    time.sleep(1)
elif end_mon == '5' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[5]/a').click()
    time.sleep(1)
elif end_mon == '6' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[6]/a').click()
    time.sleep(1)
elif end_mon == '7' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[7]/a').click()
    time.sleep(1)
elif end_mon == '8' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[8]/a').click()
    time.sleep(1)
elif end_mon == '9' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[9]/a').click()
    time.sleep(1)
else :
    driver.find_element_by_link_text(end_mon).click()
    time.sleep(1)

#종료일의 값도 xpath 값을 사용합니다.
if end_day == '1' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[1]/a').click()
    time.sleep(1)                 
elif end_day == '2' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[2]/a').click()
    time.sleep(1)
elif end_day == '3' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[3]/a').click()
    time.sleep(1)
elif end_day == '4' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[4]/a').click()
    time.sleep(1)
elif end_day == '5' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[5]/a').click()
    time.sleep(1)
elif end_day == '6' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[6]/a').click()
    time.sleep(1)
elif end_day == '7' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[7]/a').click()
    time.sleep(1)
elif end_day == '8' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[8]/a').click()
    time.sleep(1)
elif end_day == '9' :
    driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[9]/a').click()
    time.sleep(1)
else :
    driver.find_element_by_link_text(end_day).click()
    time.sleep(1)

# 날짜 입력 후 적용 버튼 클릭        
driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[3]/div/div[2]/div[3]/button').click()
time.sleep(1)

# Step 5. 검색 요청 건수만큼 화면 스크롤링 하기
# 자동 스크롤다운 함수
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(3)

i = 1
while (i <= page_cnt+2):
    scroll_down(driver) 
    i += 1
    print('%s 페이지 정보를 추출하고 있으니 잠시만 기다려 주세요~~^^' %i)

# Step 6. 현재 조회된 목록에서 URL 주소를 추출하여 리스트 생성하기
url_all_list=[]    #조회할 블로그의 URL 정보 저장용 리스트

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
url_list_1 = soup.find('ul','lst_total').find_all('li')

for a in url_list_1 :
    url_all_list.append( a.find('div','total_area').find_all('a') )

url_detail=[]
for b in range(0,len(url_all_list)) :
    url_detail.append( url_all_list[b][5]['href'] )

url_final_list=[]   # 수집할 블로그 리스트를 저장할 변수
no = 1
for c in url_detail :    
    if c.split('/')[2] == 'blog.naver.com' :
        url_final_list.append(c)
        no += 1

        if no > cnt :
            break

d_no = 1
print('')
print('정보를 수집할 블로그 URL 주소는 아래와 같습니다~~~')
print('')
for d in url_final_list :   
    print(d_no,':',d)
    d_no += 1

driver.close()
print('')

#Step 7. 수집된 URL 주소에 접속하여 데이터 추출하기
import chromedriver_autoinstaller
chromedriver_autoinstaller.install()
driver= webdriver.Chrome()

blog_addr2 = []
w_name2 = []
w_date2 = []
blog_txt2 = []

# driver.delete_all_cookies()
# time.sleep(2)

no = 1   # 전체 게시글 번호용 변수

for blog_addr in url_final_list :            
    driver.get(blog_addr)
    time.sleep(3)

    driver.switch_to.frame('mainFrame')
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    addr_1 = soup.select('#postViewArea')
    addr_2 = soup.select('div[class="se_component_wrap sect_dsc __se_component_area"]')
    addr_3 = soup.select('div[class="se-main-container"]')

    if addr_1 :
        print("%s번째 게시글 정보를 수집합니다~~~~~~~~~~" %no)
        f = open(ft_name, 'a',encoding='UTF-8')
        
        # 블로그 URL 주소
        print("1.블로그주소: ",blog_addr)
        blog_addr2.append(blog_addr)
        f.write("1.블로그 주소:"+ blog_addr + "\n")
        
        # 작성자 닉네임
        writer = soup.select("div.blog2_container > span.writer")
        try :
                wname = writer[0].get_text( )   # 작성자 닉네임
        except IndexError :
                wname = "작성자 닉네임이 없습니다"
        else :
                wname = wname.replace("\n","")

        print("2.작성자 닉네임: ",wname )
        w_name2.append(wname)
        f.write("2.작성자 닉네임:" + wname + "\n")
        
        # 작성일자
        wdate = soup.select('p[class="date fil5 pcol2 _postAddDate"]')
        try : 
                wdate = wdate[0].get_text( )
        except IndexError :
                wdate = ' '

        print("3.작성일자:",wdate)
        w_date2.append(wdate)
        f.write("3.작성 일자:" + wdate + "\n")

        # 블로그 본문 내용
        for i in addr_1:
                blog_txt = i.text.replace("\n","")
                print("4.블로그내용: \n",blog_txt) # 엔터키를 제거하는 코드
                print("\n")
                blog_txt2.append(blog_txt)
                f.write("4.블로그 내용:" + blog_txt+"\n"+"\n")

        f.close()
        no += 1

    elif addr_2 :
        print("%s번째 게시글 정보를 수집합니다~~~~~~~~~~" %no)
        f = open(ft_name, 'a',encoding='UTF-8')
        
        # 블로그 주소
        print("1.블로그주소: ",blog_addr)
        blog_addr2.append(blog_addr)
        f.write("1.블로그 주소:"+ blog_addr + "\n")
        
        # 작성자 닉네임
        writer = soup.select("div.blog2_container > span.writer")
        try :
                wname = writer[0].get_text( )   # 작성자 닉네임
        except IndexError :
                wname = "작성자 닉네임이 없습니다"
        else :
                wname = wname.replace("\n","")

        print("2.작성자 닉네임: ",wname )
        w_name2.append(wname)
        f.write("2.작성자 닉네임:" + wname + "\n")
        
        # 작성일자
        wdate = soup.select("div.blog2_container > span.se_publishDate.pcol2")
        try : 
                wdate = wdate[0].get_text( )
        except IndexError :
                wdate = ' '

        print("3.작성일자:",wdate)
        w_date2.append(wdate)
        f.write("3.작성 일자:" + wdate + "\n")

        # 블로그 본문 내용
        for i in addr_2:
                blog_txt = i.text.replace("\n","")
                print("4.블로그내용: \n",blog_txt) 
                print("\n")
                blog_txt2.append(blog_txt)
                f.write("4.블로그 내용:" + blog_txt+"\n"+"\n")

        f.close( )
        no += 1
            
    elif addr_3 :
        print("%s번째 게시글 정보를 수집합니다~~~~~~~~~~" %no)
        f = open(ft_name, 'a',encoding='UTF-8')
        
        # 블로그 주소
        print("1.블로그주소: ",blog_addr)
        blog_addr2.append(blog_addr)
        f.write("1.블로그 주소:"+ blog_addr + "\n")
        
        # 작성자 닉네임
        writer = soup.select("div.blog2_container > span.writer")
        try :
                wname = writer[0].get_text( )   # 작성자 닉네임
        except IndexError :
                wname = "작성자 닉네임이 없습니다"
        else :
                wname = wname.replace("\n","")

        print("2.작성자 닉네임: ",wname )
        w_name2.append(wname)
        f.write("2.작성자 닉네임:" + wname + "\n")
        
        # 작성일자
        wdate = soup.select("div.blog2_container > span.se_publishDate.pcol2")
        try : 
                wdate = wdate[0].get_text( )
        except IndexError :
                wdate = ' '

        print("3.작성일자:",wdate)
        w_date2.append(wdate)
        f.write("3.작성 일자:" + wdate + "\n")
        
        #블로그 본문 내용
        for i in addr_3:
                blog_txt = i.text.replace("\n","")
                print("4.블로그내용: \n",blog_txt) 
                print("\n")
                blog_txt2.append(blog_txt)
                f.write("4.블로그 내용:" + blog_txt+"\n"+"\n")

        f.close( )
        no += 1
                   
#Step 8. xls 형태와 csv 형태로 저장하기
naver_blog = pd.DataFrame()
naver_blog['블로그주소']=blog_addr2
naver_blog['작성자닉네임']=w_name2
naver_blog['작성일자']=w_date2
naver_blog['블로그내용']=blog_txt2

# csv 형태로 저장하기
naver_blog.to_csv(fc_name,encoding="utf-8-sig",index=False)

# 엑셀 형태로 저장하기
naver_blog.to_excel(fx_name ,index=False)

e_time = time.time( )
t_time = e_time - s_time


print("\n") 
print("=" *80)
print("총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("파일 저장 완료: txt 파일명 : %s " %ft_name)
print("파일 저장 완료: csv 파일명 : %s " %fc_name)
print("파일 저장 완료: xls 파일명 : %s " %fx_name)
print("=" *80)

driver.close( )
```
